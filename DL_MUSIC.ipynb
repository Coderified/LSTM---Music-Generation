```python
import os
import numpy as np

#path to your midi file set
path='D:/midi/'

#for filenames ending with ".mid" , we import the filename onto a list
files=[i for i in os.listdir(path) if i.endswith(".mid")]
```

```python
#let us define a function that reads the midi files


def read_midi(file):
    notes=[]
    notes_to_parse=None
    
#Class named converter is used
# parse method of the converter class is used to read the file
    
    midi=converter.parse(file)

# music21 supports lots of instruments    

# s2 has all instruments in midi file
    
    s2=instrument.partitionByInstrument(midi)
 

    for instruments in s2.parts:

# piano is our instrument of interest        
        if "Piano" in str(instruments):
            
            notes_to_parse = instruments.recurse()

#saves all the notes/chords in the midi file

            for element in notes_to_parse:
                
                if isinstance(element, note.Note):
                    notes.append(str(element.pitch))
                
#  chords is mix of notes
                
                elif isinstance(element,chord.Chord):
                    notes.append('.'.join(str(x) for x in element.normalOrder ))
    
    return np.array(notes) 
```

```python
# we now prep ip and op for LSTM


no_of_timesteps=32
x=[]
y=[]

for elements in new_music:
    for i in range(0,len(elements)-(no_of_timesteps),1):
        
        #ip and op preparation
        ip_swaras=elements[i:i+no_of_timesteps]
        op_swaras=elements[i+no_of_timesteps]
        
        x.append(ip_swaras)
        y.append(op_swaras)

n_patterns=len(x) 


x=np.array(x)

# np array is the preffred ip format

x.shape

y=np.array(y)
```

```python
# x_seq is not in the shape lstm takes in
# reshape it acdng to the reqrd format

x_seq=np.reshape(x_seq,(len(x_seq),no_of_timesteps,1))
```

```python
# splitting into train ann test

from sklearn.model_selection import train_test_split
```

```python
x_tr, x_val, y_tr, y_val = train_test_split(x_seq, y_seq, test_size=0.35, random_state=101)
```

```python
from keras.layers import *
from keras.models import *
from keras.callbacks import *
import keras.backend as K

model = Sequential()
model.add(LSTM(128,input_shape=(32,1),return_sequences=True))
model.add(LSTM(128))
model.add(Dense(256))
model.add(Activation('relu'))
model.add(Dense(1000))
model.add(Activation('softmax'))
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')
```

```python
history = model.fit(np.array(x_tr),np.array(y_tr),batch_size=128,epochs=30,
                    validation_data=(np.array(x_val),np.array(y_val)),verbose=1)
```
